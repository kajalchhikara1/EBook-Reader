Removing entries corresponing to invalid values

```python
invalid_values = ['Not Connect', 'I/O Timeout','Scan Timeout',"Configure",'Comm Fail','Unit Down']
mask = dataSet.isin(invalid_values).any(axis=1)

# Filter out the rows with invalid input type
data_cleaned = dataSet[~mask]
```

Converting data types of temperature and draft

```python
# Convert numerical columns to float
cleanData=data_cleaned.copy()

numerical_columns = [
    'Cyclone_Inlet_Gas_Temp',
    'Cyclone_Material_Temp',
    'Cyclone_Outlet_Gas_draft',
    'Cyclone_cone_draft',
    'Cyclone_Gas_Outlet_Temp',
    'Cyclone_Inlet_Draft'
]

for col in numerical_columns:

    cleanData[col] = pd.to_numeric(cleanData[col])
```

```python
# Set 'time' as the index
cleanData.set_index('time', inplace=True)

# Resample data by day and calculate the mean for each variable
daily_data = cleanData.resample('D').mean()
```

```python
# List of variables to plot
variables = ['Cyclone_Inlet_Gas_Temp', 'Cyclone_Material_Temp',
             'Cyclone_Outlet_Gas_draft', 'Cyclone_cone_draft',
             'Cyclone_Gas_Outlet_Temp', 'Cyclone_Inlet_Draft']

# Create a figure with subplots
fig, axes = plt.subplots(nrows=len(variables), ncols=1, figsize=(14, 4 * len(variables)))

# Create a separate bar chart for each variable
for ax, variable in zip(axes, variables):
    daily_data[variable].plot(kind='hist', ax=ax, color='skyblue', alpha=0.75)
    ax.set_title(f'Daily Average of {variable}')
    ax.set_ylabel(variable)
    ax.tick_params(axis='x', rotation=45)

    # Remove x-tick labels
    ax.set_xticklabels([])

plt.tight_layout()  # Adjusts subplot parameters for better spacing
plt.show()
```

```python
# Create new features for temperature and draft
cleanData['temperature'] = cleanData[['Cyclone_Inlet_Gas_Temp', 'Cyclone_Material_Temp', 'Cyclone_Gas_Outlet_Temp']].mean(axis=1)
cleanData['draft'] = cleanData[['Cyclone_Outlet_Gas_draft', 'Cyclone_cone_draft', 'Cyclone_Inlet_Draft']].mean(axis=1)

# Prepare the data for anomaly detection
X = cleanData[['temperature', 'draft']].dropna()
```

```python
from sklearn import svm
model = svm.OneClassSVM(kernel='rbf', gamma='auto', nu=0.05)
model.fit(X)
# Predict anomalies
predictions = model.predict(X)
cleanData['anomaly'] = np.where(predictions == -1, 1, 0)  # 1 for anomaly, 0 for normal
```

```python
# Plotting the new features and highlighting anomalies
plt.figure(figsize=(14, 8))

# Plot Temperature
plt.subplot(2, 1, 1)
plt.plot(cleanData.index, cleanData['temperature'], color='blue', label='Temperature', alpha=0.5)
plt.scatter(cleanData[cleanData['anomaly'] == 1].index, cleanData['temperature'][cleanData['anomaly'] == 1],
            color='red', label='Anomaly', marker='o', s=50)
plt.title('Temperature Over Time with Anomalies')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.legend()
plt.grid()

# Plot Draft
plt.subplot(2, 1, 2)
plt.plot(cleanData.index, cleanData['draft'], color='green', label='Draft', alpha=0.5)
plt.scatter(cleanData[cleanData['anomaly'] == 1].index, cleanData['draft'][cleanData['anomaly'] == 1],
            color='red', label='Anomaly', marker='o', s=50)
plt.title('Draft Over Time with Anomalies')
plt.xlabel('Time')
plt.ylabel('Draft')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
